```python
#Data Analysis and Visulaization
```

```python
canceled_perc = dataset['is_canceled'].value_counts(normalize = True)
print(canceled_perc)


plt.figure(figsize = (5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled' , 'Canceled'],dataset['is_canceled'].value_counts(), edgecolor = 'y', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = dataset, palette = 'Blues')
legend_labels,_ =ax1. get_legend_handles_labels()
plt.title( 'Reservation status in different hotels', size = 20)
plt.xlabel('Hotel')
plt.ylabel( "number of reservations")
plt.show()
```

```python
resort_hotel = dataset[dataset['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = dataset[dataset['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title( "Average Daily Rate in City and Resort Hotel", fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = ("Resort Hotel"))
plt.plot(city_hotel.index, city_hotel['adr'], label = ("City Hotel"))
plt.legend(fontsize = 20)
plt.show()
```

```python
dataset['month'] = dataset['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = dataset, palette = 'bright')
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title( 'Reservation status per month', size = 20)
plt.xlabel( 'month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
plt.figure(figsize=(15, 8))
plt.title('ADR per Month', fontsize=30)
data = dataset[dataset['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month', y='adr', data=data)
plt.legend(fontsize=20)
plt.show()
```

```python
cancelled_data = dataset[dataset['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 Countries With Reservation Cancelled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
dataset['market_segment'].value_counts()
```

```python
dataset['market_segment'].value_counts(normalize = True)
```

```python
cancelled_data['market_segment'].value_counts(normalize = True)
```

```python
cancelled_dataset_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_dataset_adr.reset_index(inplace=True)
cancelled_dataset_adr.sort_values('reservation_status_date', inplace=True)

dataset_not_canceled = dataset[dataset['is_canceled'] == 0]
dataset_not_canceled_adr = dataset_not_canceled.groupby('reservation_status_date')[['adr']].mean()
dataset_not_canceled_adr.reset_index(inplace=True)
dataset_not_canceled_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (18,12))
plt.plot(dataset_not_canceled_adr['reservation_status_date'],dataset_not_canceled_adr['adr'],label='Not Canceled')
plt.plot(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Canceled')
plt.title('Average Daily Rate', fontsize=30)
plt.legend()
plt.show()
```

```python
## plot the graph start from year 2016
cancelled_dataset = dataset[(dataset['is_canceled'] == 1) & (dataset['reservation_status_date'].dt.year>=2016) & (dataset['reservation_status_date'].dt.year<2017)]
cancelled_dataset_adr = cancelled_dataset.groupby('reservation_status_date')[['adr']].mean()
cancelled_dataset_adr.reset_index(inplace=True)
cancelled_dataset_adr.sort_values('reservation_status_date', inplace=True)

data_not_canceled = dataset[(dataset['is_canceled'] == 0) & (dataset['reservation_status_date'].dt.year>=2016) & (dataset['reservation_status_date'].dt.year<2017)]
data_not_canceled_adr = data_not_canceled.groupby('reservation_status_date')[['adr']].mean()
data_not_canceled_adr.reset_index(inplace=True)
data_not_canceled_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (18,12))
plt.plot(data_not_canceled_adr['reservation_status_date'],data_not_canceled_adr['adr'],label='Not Canceled')
plt.plot(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Canceled')
plt.title('Average ADR Rate for year 2016', fontsize=30)
plt.xlabel('Date')
plt.ylabel('Average ADR')
plt.legend()
plt.show()
```

```python
## plot the graph start from year 2017
cancelled_dataset = dataset[(dataset['is_canceled'] == 1) & (dataset['reservation_status_date'].dt.year>=2017)]
cancelled_dataset_adr = cancelled_dataset.groupby('reservation_status_date')[['adr']].mean()
cancelled_dataset_adr.reset_index(inplace=True)
cancelled_dataset_adr.sort_values('reservation_status_date', inplace=True)

data_not_canceled = dataset[(dataset['is_canceled'] == 0) & (dataset['reservation_status_date'].dt.year>=2017)]
data_not_canceled_adr = dataset_not_canceled.groupby('reservation_status_date')[['adr']].mean()
data_not_canceled_adr.reset_index(inplace=True)
data_not_canceled_adr.sort_values('reservation_status_date', inplace=True)

plt.figure(figsize = (18,12))
plt.plot(dataset_not_canceled_adr['reservation_status_date'],dataset_not_canceled_adr['adr'],label='Not Canceled')
plt.plot(cancelled_dataset_adr['reservation_status_date'],cancelled_dataset_adr['adr'],label='Canceled')
plt.title('Average ADR Rate for year 2017', fontsize=30)
plt.xlabel('Date')
plt.ylabel('Average ADR')
plt.legend()
plt.show()
```
